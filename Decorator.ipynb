```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper
@my_decorator    #and here u can add more than one decorator each one in a line
def say_whee():
    print("Whee!")

#say_whee = my_decorator(say_whee)
```

```python
t=0
def perm(l):
    if len(l) <= 1:
        return [l]
    r = []
    for i in range(len(l)):
        s = l[:i] + l[i+1:]
        p = perm(s)
        for x in p:
            r.append(l[i:i+1] + x)
            global t
            t+=1
    print(t)
    return r
x= [3,4,5,6,7,8,9]
perm(x)
```
